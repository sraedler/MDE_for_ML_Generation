# Import section
Imports for notebook

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import datetime
```

<div>
<h1>Dataset load</h1>

<p>Load the dataset(s)</p>
</div>


```python
path = "weather_split_complex_one.csv"
separator = ","
enc = "UTF-8"
df_one = pd.read_csv(path, sep=separator, encoding=enc, skiprows=0)
```

```python
df_two = pd.read_csv("weather_split_complex_two.csv", sep=";", encoding="UTF-8", skiprows=0)
```

<div>
<h1>Data Preprocessing: Date Conversion&nbsp;</h1>
</div>

<p>&nbsp;</p>


```python
df_one["date"] = df_one["date"].apply(lambda old_date, old_format, new_format: datetime.datetime.strptime(old_date, old_format).strftime(new_format), args=("%d-%m-%Y", "%Y-%m-%d"))
```

<div>
<h1>Dataframe Merge</h1>

<p>Merge dataframes</p>
</div>


```python
Merge_DF = pd.merge(left=df_one, right=df_two, left_on="date", right_on="date_date", how="inner")
```

<div>
<h1>Data Pre-Processing: Label Encoding</h1>
</div>


```python
Merge_DF["weather"]=LabelEncoder().fit_transform(Merge_DF["weather"])
```

```python
X=Merge_DF[["precipitation", "temp_max", "temp_min", "wind"]]
y=Merge_DF.weather
TrainSplit_train_X, TrainSplit_test_X, TrainSplit_train_y, TrainSplit_test_y = train_test_split(X, y,random_state = 0, train_size=0.7)
```

<div>
<h2>Decision Tree Regressor</h2>

<p>Creation of decision tree regressor</p>
</div>


```python
mod_ML_2=DecisionTreeRegressor(random_state=1)
mod_ML_2.fit(TrainSplit_train_X, TrainSplit_train_y)
```

<div>
<div>
<h1>Create Model</h1>

<p>Section where models are created</p>
</div>

<h2>Random Forest Regressor</h2>
</div>


```python
mod_ML_1=RandomForestRegressor(random_state=1)
mod_ML_1.fit(TrainSplit_train_X, TrainSplit_train_y)
```

<div>
<h1>Prediction</h1>

<p>Section where models are used to predict</p>
</div>

<div>
<h2>Random&nbsp;Forest Regressor</h2>
</div>

<p>&nbsp;</p>


```python
pre_ML_1 = mod_ML_1.predict(TrainSplit_test_X)
```

<div>
<h2>Decision Tree Regressor</h2>
</div>


```python
pre_ML_2 = mod_ML_2.predict(TrainSplit_test_X)
```

<div>
<h1>Metrics</h1>

<p>Section for metrics for the used model(s)</p>
</div>

<div>
<h2>Random Forest Regressor</h2>
</div>

<p>&nbsp;</p>


```python
mae = mean_absolute_error(TrainSplit_test_y, pre_ML_1)
print("Mean Absolute Error: %f" % mae)
print("First mean absolute error")
```

<div>
<h2>Decision Tree Regressor</h2>
</div>


```python
mae = mean_absolute_error(TrainSplit_test_y, pre_ML_2)
print("Mean Absolute Error: %f" % mae)
print("Second mean absolute error")
```
